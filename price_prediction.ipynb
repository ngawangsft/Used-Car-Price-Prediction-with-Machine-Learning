```python
# Exploring Power Column Further
q95 = df_clean['Power'].quantile(0.95)
q99 = df_clean['Power'].quantile(0.99)
q999= df_clean['Power'].quantile(0.999)

print(f"95th percentile: {q95:.2f}")
print(f"99th percentile: {q99:.2f}")
print(f"99.9th percentile horsepower: {q999:.2f}") 
print()

# Lets find out our IQR 
p = df_clean['Power']
p_pos = p[p > 0] 
lp = np.log1p(p_pos) 

Q1, Q3 = lp.quantile(0.25), lp.quantile(0.75)
IQR = Q3 - Q1
upper_lp = Q3 + 1.5 * IQR
upper_power = np.expm1(upper_lp)

# Lets use a little higher than our IQR 
upper_cutoff = max(upper_power, 450)

print(f"IQR-based upper Power cutoff: {upper_power:.2f} HP")
print(f"Chosen cutoff for cleaning: {upper_cutoff:.2f} HP\n")
```

```python
# Clean up our data 
original_size= len(df_clean)
df_clean= df_clean[df_clean['Power'] <= upper_cutoff]
new_size= len(df_clean)
removed= original_size - new_size

print("After applying upper Power cutoff:")
print("Original dataset size:", original_size)
print("New dataset size:", new_size)
print("Removed rows:", removed)
print(f"Percentage removed: {removed/original_size:.2%}\n")

# Remove cars with zero horsepower
original_size= len(df_clean)
df_clean= df_clean[df_clean['Power'] > 0]
new_size= len(df_clean)

print(f"Removed {original_size - new_size} cars with 0 HP")
print(f"Remaining dataset size: {new_size}")
```

## Linear Regression Baseline Sanity Check 
---

```python
# State Target and Features
target= 'Price'

numerical_features= [
    'Power', 
    'Mileage', 
    'CarAge', 
    'RegistrationMonth', 
    'DaysSinceListing'
]

categorical_features= [
    'VehicleType', 
    'Gearbox', 
    'Model', 
    'FuelType', 
    'Brand', 
    'NotRepaired'
] 
# For first run, dont include PostalCode, we can to test it out later to see if RMSE increases with it included
# Re-ran second baseline with PostalCode included, worse RMSE. So remove it 

X = df_clean[numerical_features + categorical_features]
y = df_clean[target]

X_train,X_test,y_train, y_test= train_test_split(
    X, y, test_size=0.25, random_state=12345)

# Preprocessing
preprocessor = ColumnTransformer([
    ('num', StandardScaler(), numerical_features),
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
])

# Linear Regression Pipeline
lr_pipeline = Pipeline([
    ('preprocessing', preprocessor),
    ('model', LinearRegression())
])
```

```python
# Measure training time
start_train = time.time()
lr_pipeline.fit(X_train, y_train)
end_train = time.time()
train_time = end_train - start_train

# Measure prediction speed
start_pred = time.time()
y_pred = lr_pipeline.predict(X_test)
end_pred = time.time()
pred_time = end_pred - start_pred

# Get Raw RMSE
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

# Get relative RMSE
mean_price = y_train.mean()
rmse_ratio = rmse / mean_price

print(f"Linear Regression Raw RMSE: {rmse:.2f}")
print(f"Relative RMSE: {rmse_ratio*100:.1f}% of mean car price")
print(f"Training Time: {train_time:.2f} seconds")
print(f"Prediction Speed (test set): {pred_time:.4f} seconds")
```

```python
# Checking the size of your categorical features
# Since our Model and Brand has too many unqiue values;
# we'll have to make a smaller sample before we model our RandomForest
for col in categorical_features:
    unique_count = df_clean[col].nunique()
    print(f"{col}: {unique_count} unique values")

print(f"\nDataset size: {len(X_train)} training samples")
```

## RandomForest Model (100k Sample Size with HyperTuning/ Grid Search)
---

```python
# Pipeline for Random Forest
rf_pipe = Pipeline([
    ('preprocessing', preprocessor),
    ('model', RandomForestRegressor(random_state=12345, n_jobs=-1))
])

# Setting up hyperparameters grid
param_grid = {
    'model__n_estimators': [30, 40, 50],
    'model__max_depth': [10, 15, 20]
}

grid_search = GridSearchCV(
    estimator=rf_pipe,
    param_grid=param_grid,
    cv=3,
    scoring='neg_root_mean_squared_error',
    verbose=2
)

# Adjusting Sample Size 
sample_size = 100000
X_train_sample = X_train.sample(n=sample_size, random_state=12345)
y_train_sample = y_train.loc[X_train_sample.index]

# Run grid search
start_time = time.time()
grid_search.fit(X_train_sample, y_train_sample)
end_time = time.time()

print(f"Grid search time: {end_time - start_time:.2f} seconds")
print("Best Parameters:", grid_search.best_params_)
print("Best RMSE:", -grid_search.best_score_)
```

```python
# 100k Random Forest test
sample_size = 100000
X_train_sample = X_train.sample(n=sample_size, random_state=12345)
y_train_sample = y_train.loc[X_train_sample.index]

print(f"Original training size: {len(X_train):,} samples")
print(f"Sample training size: {len(X_train_sample):,} samples")
print(f"Reduction: {(1 - len(X_train_sample)/len(X_train))*100:.1f}%")
print(f"Test set remains: {len(X_test):,} samples")
```

```python
# Random Forest Pipeline (100k using best Parameters in GridSearchCV)
rf_pipe = Pipeline([
    ('preprocessing', preprocessor),
    ('model', RandomForestRegressor(
        n_estimators=50,
        max_depth=20,
        random_state=12345,
        n_jobs=-1
    ))
])

# Measure training time
start_train = time.time()
rf_pipe.fit(X_train_sample, y_train_sample)
end_train = time.time()
train_time = end_train - start_train

# Measure prediction speed
start_pred = time.time()
y_pred = rf_pipe.predict(X_test)
end_pred = time.time()
pred_time = end_pred - start_pred

# Compute RMSE
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

# Compute relative RMSE (compared to sample mean)
mean_price = y_train_sample.mean()
rmse_ratio = rmse / mean_price

print(f"\nRandom Forest (100k sample) Raw RMSE: {rmse:.2f}")
print(f"Relative RMSE: {rmse_ratio*100:.1f}% of mean car price")
print(f"Training Time: {train_time:.2f} seconds")
print(f"Prediction Speed (test set): {pred_time:.4f} seconds")
```

## LightGBM Pipeline

```python
# Split into training + test
X_train_full, X_test, y_train_full, y_test = train_test_split(
    X, y, test_size=0.25, random_state=12345
)

# Split training into training + validation
X_train, X_val, y_train, y_val = train_test_split(
    X_train_full, y_train_full, test_size=0.2, random_state=12345
)

# Create copies for LightGMB
X_train_lgb = X_train.copy()
X_val_lgb   = X_val.copy()
X_test_lgb  = X_test.copy()

label_encoders = {}
for col in categorical_features:
    le = LabelEncoder()
    combined_data = pd.concat([X_train_lgb[col], X_val_lgb[col], X_test_lgb[col]])
    le.fit(combined_data)
    
    X_train_lgb[col] = le.transform(X_train_lgb[col])
    X_val_lgb[col]   = le.transform(X_val_lgb[col])
    X_test_lgb[col]  = le.transform(X_test_lgb[col])
    label_encoders[col] = le

categorical_indices = [X_train_lgb.columns.get_loc(col) for col in categorical_features]

train_data = lgb.Dataset(X_train_lgb, label=y_train, categorical_feature=categorical_indices)
val_data   = lgb.Dataset(X_val_lgb, label=y_val, categorical_feature=categorical_indices, reference=train_data)
```

```python
params = {
    'objective': 'regression',
    'metric': 'rmse',
    'boosting_type': 'gbdt',
    'learning_rate': 0.075,
    'num_leaves': 42,
    'max_depth': -1,
    'verbose': -1,
    'n_jobs': -1
}

start_train = time.time()
lgb_model = lgb.train(
    params,
    train_data,
    num_boost_round=1500,
    valid_sets=[val_data],
    early_stopping_rounds=50,
    verbose_eval=100
)
end_train = time.time()
train_time = end_train - start_train

start_pred = time.time()
y_pred = lgb_model.predict(X_test_lgb, num_iteration=lgb_model.best_iteration)
end_pred = time.time()
pred_time = end_pred - start_pred

rmse = mean_squared_error(y_test, y_pred, squared=False)
rmse_ratio = rmse / y_train_full.mean()

print(f"\nLightGBM Raw RMSE: {rmse:.2f}")
print(f"Relative RMSE: {rmse_ratio*100:.1f}% of mean car price")
print(f"Training Time: {train_time:.2f} seconds")
print(f"Prediction Speed (test set): {pred_time:.4f} seconds")

```
